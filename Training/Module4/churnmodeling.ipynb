```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
# df.iloc[:,0]
X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
geo_dummies = pd.get_dummies(X['Geography'],drop_first=True).astype('int64')
gender_dummies = pd.get_dummies(X['Gender'],drop_first=True).astype('int64')
```

```python
X = pd.concat([X,geo_dummies,gender_dummies],axis = 1 )
```

```python
X.drop(columns=['Geography','Gender'],inplace=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# feature scaling 
# we should try to scale down all the features to a standard term or the same scale so that it is 
# easy for coputation in the next layers it doesnot take a long time and the backprop is also done 
# convergence will happen quickly 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras 
from keras.models import Sequential #responsible for all the learnings 
from keras.layers import Dense 
from keras.layers import ReLU,LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units=6 , kernel_initializer = 'he_uniform', activation='relu',input_dim = 11))  #first layer has 11 input neurons and second layer has 6 neurons 
classifier.add(Dense(units=6 , kernel_initializer = 'he_uniform', activation='relu')) # added 6 more 
classifier.add(Dense(units=1 , kernel_initializer = 'glorot_uniform', activation='sigmoid')) # output layer
```

```python
# compiling the ANN 
classifier.compile(optimizer='adam',loss ='binary_crossentropy',metrics=['accuracy'])
# whenever there is binary classification we should use binary_crossentropy, if there are multiple categories use categorical_crossentropy.
```

```python
model_history = classifier.fit(X_train, y_train, validation_split=0.33, batch_size= 10, epochs = 100)
```
